```python
import pandas as pd
import glob

# Define the path and pattern to search for files
path = '/Users/sofialozano/Library/CloudStorage/OneDrive-EmoryUniversity/EMORY FALL 2024/ECON 521 POLICY CAUSAL INFERENCE/Final Project 521/521 Prices Amazon keepa'
pattern = f"{path}/*NEW*.csv"

# Find all files matching the pattern
files = glob.glob(pattern)

# Read each file into a dataframe and store them in a dictionary
dataframes = {file: pd.read_csv(file) for file in files}

# Print the names of the files imported
print("Imported files:", list(dataframes.keys()))
```

```python
# Create a dictionary to store the dataframes with the new names
named_dataframes = {}

# Iterate over the dataframes dictionary
for file_path, df in dataframes.items():
    # Extract the name from the file path
    name = file_path.split('NEW_')[1].split('_amazon_asins.csv')[0]
    # Store the dataframe in the new dictionary with the extracted name
    named_dataframes[name] = df

# Print the names of the new dataframes
print("Named dataframes:", list(named_dataframes.keys()))
```

```python
for name, df in named_dataframes.items():
    print(f"DataFrame: {name}")
    print(df.head())
    print("\n")
```

```python
for name, df in named_dataframes.items():
    df['product_name'] = name

# Verify the changes
for name, df in named_dataframes.items():
    print(f"DataFrame: {name}")
    print(df.head())
    print("\n")
```

```python
#Step 3: Create df3 with food items as treatments
df3 = pd.DataFrame()
for product_name, df in named_dataframes.items():
    df = df.copy()
    df["product"] = product_name
    df3 = pd.concat([df3, df], ignore_index=True)

# Assign treatment for df3. Select the treated products: 
df3["treatment"] = df3["product"].isin(["arborio", "RolandArborioSuperfino", "BarillaPasta", "avocados", "almondmilk", "nutmix"]).astype(int)

# Drop unnecessary columns and ensure proper formatting
df3.drop(columns=["Unnamed: 0"], inplace=True, errors='ignore')  # Ignore if column doesn't exist
df3["date"] = pd.to_datetime(df3["date"])

# View the result
print("DataFrame 3 (all treatment products):")
df3
```

```python
df3['post'] = (df3['date'] > '2017-08-28').astype(int)
df3 
```

```python
# Convert the "price" column to log units for df3
import numpy as np 
df3['log_price'] = np.log(df3['price'])

# View the result
df3 
```

```python
treated_product_names = df3[df3['treatment'] == 0]['product_name'].unique()
print("Product names with treatment == 0:", treated_product_names)
```
